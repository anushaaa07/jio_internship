```python
import json
import torch
from transformers import (AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline)
```

```python
config_data=json.load(open("/content/sample_data/config.json"))
HF_TOKEN = config_data.get("HF_TOKEN")
```

```python
print(HF_TOKEN)
```

```python
model_name = "meta-llama/Meta-Llama-3-8B"
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, token=HF_TOKEN)
tokenizer.pad_token = tokenizer.eos_token
```

```python
!pip install transformers
import transformers
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", quantization_config=bnb_config, token=HF_TOKEN)
```

```python
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
def get_response(prompt):
  sequences = text_generator(prompt)
  gen_text = sequences[0]['generated_text']
  return gen_text
```

```python
prompt = "What is Machine Learning?"
```

```python
llama3_response= get_response(prompt)
```
